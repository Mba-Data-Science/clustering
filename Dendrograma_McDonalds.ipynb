```python
#Importando principais Bibliotecas
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
#Transformando a base de dados CSV em um DataFrame
lanches = pd.read_csv('C:/Users/Felipe Veras/Desktop/Aula MBA 15.06/dados/MCDONALDS.csv', sep=';', index_col = 0, decimal = ',')

ferro = lanches['Ferro'].replace(',', '.')
```

```python
#Vendo os dados
lanches
```

```python
#Fazendo o preprocessamento dos dados. Usando o Z-score para padronizar todos os dados

from sklearn import preprocessing
scaler = preprocessing.StandardScaler()

lanches_padronizado = scaler.fit_transform(lanches)

lanches_padronizado = pd.DataFrame(lanches_padronizado, columns = lanches.columns, index = lanches.index)

lanches_padronizado
```

```python
from scipy.cluster.hierarchy import dendrogram
from sklearn.cluster import AgglomerativeClustering 

## Cada modelo é um dendrograma usando um linkage diferente
model_single = AgglomerativeClustering(distance_threshold = 0, n_clusters = None, linkage = 'single')

model_single.fit(lanches_padronizado)
```

```python
#Função para plot do dendrograma
##QUEM SOUBER UM MÉTODO MAIS FÁCIL, ME DIZ POR FAVOR!
def plot_dendrogram(model, **kwargs):

    # Children of hierarchical clustering
    children = model.children_

    # Distances between each pair of children
    # Since we don't have this information, we can use a uniform one for plotting
    distance = np.arange(children.shape[0])

    # The number of observations contained in each cluster level
    no_of_observations = np.arange(2, children.shape[0]+2)

    # Create linkage matrix and then plot the dendrogram
    linkage_matrix = np.column_stack([children, distance, no_of_observations]).astype(float)

    # Plot the corresponding dendrogram
    dendrogram(linkage_matrix, **kwargs)
```

```python
#Plot do Dendrograma Single Linkage
#Como colocar o nome dos lanches no eixo x?
plt.title('Hierarchical Clustering Dendrogram - Single')
plot_dendrogram(model_single, truncate_mode='level')
plt.xlabel("Number of points in node (or index of point if no parenthesis)")
plt.show()
```

```python
#Usando o metodo do Elbow para encontrar o numero de clusters
from sklearn.cluster import KMeans

rsquared = []

for n in range(2,10):
    model = KMeans(n_clusters = n)
    model.fit(lanches_padronizado)
    rsquared.append(model.inertia_)

plt.grid(linestyle = '-')
plt.plot(range(2,10), rsquared)

```

```python
#Cortando a árvore para criar os grupos
## from scipy.cluster.hierarchy import linkage = diz o linkage usado para cortar os grupos

from scipy.cluster.hierarchy import linkage
from scipy.cluster.hierarchy import cut_tree
hc_single = linkage(lanches_padronizado, 'single')
grupo_lanches = cut_tree(hc_single,n_clusters = 4)

grupo_lanches_df = pd.DataFrame(grupo_lanches, index = lanches_padronizado.index, columns = ['Grupos'])
grupo_lanches_df
```
