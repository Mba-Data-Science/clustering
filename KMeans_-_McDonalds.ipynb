```python
#Importando principais Bibliotecas
import numpy as np
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
```

```python
#Transformando a base de dados CSV em um DataFrame
lanches = pd.read_csv('C:/Users/Felipe Veras/Desktop/Aula MBA 15.06/dados/MCDONALDS.csv', sep=';', index_col = 0, decimal = ',')
```

```python
#Observando meu DataFrame
lanches
```

```python
#Fazendo o preprocessamento dos dados. Usando o Z-score para padronizar todos os dados

from sklearn import preprocessing
scaler = preprocessing.StandardScaler()

lanches_padronizado = scaler.fit_transform(lanches)

lanches_padronizado = pd.DataFrame(lanches_padronizado, columns = lanches.columns, index = lanches.index)

lanches_padronizado
```

```python
#Fazendo o modelo usando 2 clusters e salvando em lanches['grupo'] os grupos de cada lanche
from sklearn.cluster import KMeans

model = KMeans(n_clusters = 2)
kmeans = model.fit(lanches_padronizado)

lanches['grupo'] = kmeans.labels_

kmeans.cluster_centers_
```

```python
lanches
```

```python
# Fazendo uma analise de PCA para redimensionar os dados para apenas 2 e assim conseguir visualizar o modelo KMeans

from sklearn import decomposition

m = lanches_padronizado.shape[1]
k = 2

mod_PCA = decomposition.PCA(n_components = m)
DF_PCA = pd.DataFrame(mod_PCA.fit_transform(lanches_padronizado),
                                            columns=["PC%d" %k for k in range (1,m+1)]).iloc[:, :k]
print(DF_PCA)
            
```

```python
#Plotando o modelo com 2 Clusters usando os principais componentes como eixos do modelo
colors = [{0: 'r', 1:'b', 2:'g', 3: 'y', 4:'p', 5: 'b'}[x] for x in lanches['grupo']]

plt.figure(figsize=(10,10)) 
plt.scatter(x=DF_PCA['PC1'], y=DF_PCA['PC2'], color = colors)

for n, txt in enumerate(lanches.index):
    plt.annotate(txt, (DF_PCA['PC1'][n], DF_PCA['PC2'][n]))
```

```python
# Agora vamos repetir o modelo fazendo com 3,4,5 e 6 clusters. 

model3 = KMeans(n_clusters = 3)
model4 = KMeans(n_clusters = 4)
model5 = KMeans(n_clusters = 5)
model6 = KMeans(n_clusters = 6)

kmeans3 = model3.fit(lanches_padronizado)
kmeans4 = model4.fit(lanches_padronizado)
kmeans5 = model5.fit(lanches_padronizado)
kmeans6 = model6.fit(lanches_padronizado)

lanches3 = lanches.copy(deep = True)
lanches4 = lanches.copy(deep = True)
lanches5 = lanches.copy(deep = True)
lanches6 = lanches.copy(deep = True)

lanches3['grupo'] = kmeans3.labels_
lanches4['grupo'] = kmeans4.labels_
lanches5['grupo'] = kmeans5.labels_
lanches6['grupo'] = kmeans6.labels_
```

```python
#Plotando o modelo com 3 Clusters
colors = [{0: 'r', 1:'b', 2:'g', 3: 'y', 4:'p', 5: 'b'}[x] for x in lanches3['grupo']]

#fig, ax = plt.subplots()
plt.figure(figsize=(10,10)) 
plt.scatter(x=DF_PCA['PC1'], y=DF_PCA['PC2'], color = colors)

for n, txt in enumerate(lanches3.index):
    plt.annotate(txt, (DF_PCA['PC1'][n], DF_PCA['PC2'][n]))
```

```python
#Plotando o modelo com 4 Clusters
colors = [{0: 'r', 1:'b', 2:'g', 3: 'y', 4:'p', 5: 'b'}[x] for x in lanches4['grupo']]

#fig, ax = plt.subplots()
plt.figure(figsize=(10,10)) 
plt.scatter(x=DF_PCA['PC1'], y=DF_PCA['PC2'], color = colors)

for n, txt in enumerate(lanches4.index):
    plt.annotate(txt, (DF_PCA['PC1'][n], DF_PCA['PC2'][n]))
```

```python
#Plotando o modelo com 5 Clusters
colors = [{0: 'r', 1:'b', 2:'g', 3: 'y', 4:'purple', 5: 'b'}[x] for x in lanches5['grupo']]

#fig, ax = plt.subplots()
plt.figure(figsize=(10,10)) 
plt.scatter(x=DF_PCA['PC1'], y=DF_PCA['PC2'], color = colors)

for n, txt in enumerate(lanches5.index):
    plt.annotate(txt, (DF_PCA['PC1'][n], DF_PCA['PC2'][n]))
```

```python
#Plotando o modelo com 6 Clusters
colors = [{0: 'r', 1:'b', 2:'g', 3: 'y', 4:'purple', 5: 'black'}[x] for x in lanches6['grupo']]

#fig, ax = plt.subplots()
plt.figure(figsize=(10,10)) 
plt.scatter(x=DF_PCA['PC1'], y=DF_PCA['PC2'], color = colors)

for n, txt in enumerate(lanches6.index):
    plt.annotate(txt, (DF_PCA['PC1'][n], DF_PCA['PC2'][n]))
```

```python
#Usando o metodo do Elbow para encontrar o numero de clusters

rsquared = []

for n in range(2,10):
    model_elbow = KMeans(n_clusters = n)
    model_elbow.fit(lanches_padronizado)
    rsquared.append(model_elbow.inertia_)

plt.grid(linestyle = '-')
plt.plot(range(2,10), rsquared)

#Baseado no grafico, eu escolheria 4.
```
